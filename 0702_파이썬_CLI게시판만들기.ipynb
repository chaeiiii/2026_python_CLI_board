```python

print("== CLI 게시판 실행 ==")

last_article_id = 0
articles = [] # 게시글 저장소

while True:
    cmd = input("명령어 ) ").strip()

    if cmd == "exit":
      break

    elif cmd == "article write":
      last_article_id += 1
      id = last_article_id
      title = input("제목 : ")
      body = input("내용 : ")
      article = {"id" : id, "title" : title, "body" : body}
      articles.append(article)
      print("{}번 글이 생성되었습니다.".format(last_article_id))

    elif cmd == "article list":
      if not articles:
        print("게시글이 없습니다.")
      else:
        print("=========================================")
        print("   번호      /     제목       /     내용   ")
        for a in reversed(articles):
          print("   {}      /     {}       /     {}   ".format(a["id"], a["title"], a["body"]))
        print("=========================================")
    elif "article delete" in cmd:
      cmd_num = int(cmd.split()[-1])
      for a in range(len(articles)):
        if cmd_num == articles[a]["id"]:
          del articles[a]
          print("{}번 글이 삭제되었습니다.".format(cmd_num))
          break
      else:
        print("{}번 글은 존재하지 않습니다.".format(cmd_num))

    else:
      print("지원하지 않는 기능입니다.")

print("== CLI 게시판 종료 ==")
```

```python
## 0702 수정 기능
from datetime import datetime, timezone, timedelta
print("== CLI 게시판 실행 ==")

last_article_id = 0
articles = [] # 게시글 저장소
created_dates =[]
while True:
    cmd = input("명령어 ) ").strip()

    if cmd == "exit":
      break

    elif cmd == "article write":
      last_article_id += 1
      id = last_article_id
      title = input("제목 : ")
      body = input("내용 : ")
      article = {"id" : id, "title" : title, "body" : body}
      articles.append(article)
      created_date = datetime.now(timezone(timedelta(hours=9))).strftime('%Y-%m-%d %H:%M:%S')
      created_dates.append(created_date)
      # print(created_date)
      print("{}번 글이 생성되었습니다".format(last_article_id))

    elif cmd == "article list":
      if not articles:
        print("게시글이 없습니다")
      else:
        print("=========================================")
        print("   번호      /     제목       /     내용   ")
        for a in reversed(articles):
          print("   {}      /     {}       /     {}   ".format(a["id"], a["title"], a["body"]))
        print("=========================================")

    elif "article delete" in cmd: #삭제기능
      cmdBits = cmd.split()
      if len(cmdBits) < 3:
        print("명령어를 확인해주세요(길이 모자람)")
        continue
      if not cmdBits[2].isdigit():
        print("명령어를 확인해주세요(번호대신 str 입력)")
        continue
      cmd_num = int(cmdBits[-1])

      for a in range(len(articles)):
        if cmd_num == articles[a]["id"]:
          del articles[a]
          print("{}번 글이 삭제되었습니다".format(cmd_num))
          break
      else:
        print("{}번 글은 존재하지 않습니다".format(cmd_num))

    elif "article edit" in cmd: # 수정기능
      cmdBits = cmd.split()
      if len(cmdBits) < 3:
        print("명령어를 확인해주세요(길이 모자람)")
        continue
      if not cmdBits[2].isdigit():
        print("명령어를 확인해주세요(번호대신 str 입력)")
        continue
      cmd_num = int(cmdBits[-1])

      for a in range(len(articles)):
        if cmd_num == articles[a]["id"]:
          print("기존 제목 :",articles[a]["title"])
          print("기존 내용 :",articles[a]["body"])
          edit_title = input("새 제목 : ")
          edit_body = input("새 내용 : ")
          articles[a]["title"]=edit_title
          articles[a]["body"]=edit_body

          print("{}번 글이 수정되었습니다".format(cmd_num))
          break
      else:
        print("{}번 글은 존재하지 않습니다".format(cmd_num))

     #5. 게시글 상세보기 기능 구현
    elif "article detail" in cmd:
      cmdBits = cmd.split()
      if len(cmdBits) < 3:
        print("명령어를 확인해주세요(길이 모자람)")
        continue
      if not cmdBits[2].isdigit():
        print("명령어를 확인해주세요(번호대신 str 입력)")
        continue
      cmd_num = int(cmdBits[-1])
      for a in range(len(articles)):
        if cmd_num == articles[a]["id"]:
          print("번호 :",articles[a]["id"])
          print("날짜 :",created_dates[a])
          print("제목 :",articles[a]["title"])
          print("내용 :",articles[a]["body"])

          break
      else:
        print("{}번 글은 존재하지 않습니다".format(cmd_num))
    else:
      print("지원하지 않는 기능입니다")

print("== CLI 게시판 종료 ==")
```
