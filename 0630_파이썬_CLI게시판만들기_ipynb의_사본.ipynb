```python
import numpy as np
print("==CLI 게시판 실행 ==")
i = 1
articlelist = []
# articlelist=np.array(articlelist)
while True:
  a = input("명령어 )")
  if a == 'article write':
    title = input("제목 : ")
    info = input("내용: ")
    print(str(i)+"번 글이 생성 되었습니다")

    articlelist.append([i,title,info])


    print(articlelist)
    i+=1

  elif a == 'exit':
    break
  elif a == 'article list':
    print("====================================")
    print("번호   /   제목    /   내용    ")
    print(" ",articlelist[0][0],"      ", articlelist[0][1],"     ",articlelist[0][2])

    print("====================================")
  else:
    print("지원하지 않는 기능입니다.")
print("== CLI 게시판 종료 ==")

```
